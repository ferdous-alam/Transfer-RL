#### Plot task similarity

```python
import matplotlib.pyplot as plt
import numpy as np

# latex font
plt.rcParams.update({
"text.usetex": True,
"font.family": "sans-serif",
"font.sans-serif": "Helvetica",
})


plt.rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})
plt.rc('text', usetex=True)
```

```python
def plot_similarity(*args, fig_name):
    mean, std = [], []
    for k in args:
        mean.append(np.mean(k))
        std.append(np.std(k))
    
    x_axis = [i for i in range(len(mean))]
    plt.figure(figsize=(10, 8))
    plt.errorbar(x_axis, mean, lw=5.0, yerr=std, fmt='o', ms=20,
                ecolor='black', color='black', capsize=14.0, capthick=12.0, 
                markeredgewidth=5.0)
    plt.xlabel(r'tasks', fontsize=24)
    plt.ylabel(r'$\Xi^{dyn}_{\mathcal{S}, \mathcal{T}}$', fontsize=30)
    tick_labels = ['$(\mathcal{S}, \mathcal{T}_1)$', '$(\mathcal{S}, \mathcal{T}_2)$', 
                    '$(\mathcal{S}, \mathcal{T}_3)$', '$(\mathcal{S}, \mathcal{T}_4)$']
    plt.xticks(x_axis, tick_labels, fontsize=24)    
    plt.yticks(fontsize=24)
    plt.rcParams['axes.linewidth'] = 2.0 
    plt.savefig(f'figures/{fig_name}.pdf', bbox_inches='tight', dpi=300)
```

--> Pendulum

```python
sim1 = np.load('data/pendulum/pendulum_g10_12.npy')
sim2 = np.load('data/pendulum/pendulum_g10_14.npy')
sim3 = np.load('data/pendulum/pendulum_g10_16.npy')
sim4 = np.load('data/pendulum/pendulum_g10_18.npy')

vals = [sim1, sim2, sim3, sim4]
means = np.mean(vals, 1)
print(f'means: {means}')
```

```python
plot_similarity(sim1, sim2, sim3, sim4, fig_name='pendulum_dyn_task_sim')
```

```python
sim1 = np.load('data/halfcheetah/halfcheetah_xml0_xml1.npy')
sim2 = np.load('data/halfcheetah/halfcheetah_xml0_xml2.npy')
sim3 = np.load('data/halfcheetah/halfcheetah_xml0_xml3.npy')
sim4 = np.load('data/halfcheetah/halfcheetah_xml0_xml4.npy')

plot_similarity(sim1, sim2, sim3, sim4, fig_name='halfcheetah_dyn_task_sim')
```
